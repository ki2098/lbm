```python
import numpy as np
import sympy as sp
import math
import itertools as it

Q = 27
D = 3

def cmk(i,j,k):
    return i*D*D + j*D + k

def pdf(i,j,k):
    return (i + 1)*D*D + (j + 1)*D + (k + 1)

def cmk2str(i,j,k):
    return f'_{i}{j}{k}'

def pdf2str(i,j,k):
    name = 'orl'
    return f'_{name[i]}{name[j]}{name[k]}'
```

```python
for a,b,c in it.product(range(-1,2), repeat=3):
    print(f'const Int {pdf2str(a,b,c)} = {pdf(a,b,c)};')
```

```python
for a,b,c in it.product(range(0,3), repeat=3):
    print(f'const Int {cmk2str(a,b,c)} = {cmk(a,b,c)};')
```

```python
f = np.ndarray(Q, dtype=object)
for a,b,c in it.product(range(-1,2), repeat=3):
    f[pdf(a,b,c)] = sp.Symbol(f'f[{pdf2str(a,b,c)}]')
f
```

```python
m = np.ndarray(Q, dtype=object)
for a,b,c in it.product(range(0,3), repeat=3):
    m[cmk(a,b,c)] = sp.Symbol(f'm[{cmk2str(a,b,c)}]')
m
```

```python
k = np.ndarray(Q, dtype=object)
for a,b,c in it.product(range(0,3), repeat=3):
    k[cmk(a,b,c)] = sp.Symbol(f'k[{cmk2str(a,b,c)}]')
k
```

# 1. $\bm m \leftrightarrow \bm f$

$\bm m = M\bm f$

```python
M = np.ndarray((Q,Q), dtype=int)
for a,b,c in it.product(range(0,3), repeat=3):
    for x,y,z in it.product(range(-1,2), repeat=3):
        row = cmk(a,b,c)
        col = pdf(x,y,z)
        M[row, col] = (x**a) * (y**b) * (z**c)
M
```

```python
tmp = M@f
for a,b,c in it.product(range(0,3), repeat=3):
    row = cmk(a,b,c)
    print(f'{m[row]} = {tmp[row]};')
```

$\bm f = M^{-1}\bm m$

```python
Mi = np.linalg.inv(M)
Mi
```

```python
tmp = sp.factor(Mi@m)
for a,b,c in it.product(range(-1,2), repeat=3):
    row = pdf(a,b,c)
    print(f'{f[row]} = {tmp[row]};')
```

# 2. $\bm\kappa \leftrightarrow \bm m$

$\bm\kappa = K\bm m$

```python
K = np.ndarray((Q,Q), dtype=object)
u,v,w = sp.symbols('u,v,w')
for a,b,c in it.product(range(0,3), repeat=3):
    for x,y,z in it.product(range(0,3), repeat=3):
        row = cmk(a,b,c)
        col = cmk(x,y,z)
        if x<=a and y<=b and z<=c:
            K[row, col] = math.comb(a,x)*math.comb(b,y)*math.comb(c,z)*((-u)**(a-x))*((-v)**(b-y))*((-w)**(c-z))
        else:
            K[row, col] = 0

K = sp.Matrix(K)
K
```

```python
tmp = K@m
for q in range(Q):
    tmp[q] = sp.simplify(sp.collect(sp.collect(tmp[q], u), v))
for a,b,c in it.product(range(0,3), repeat=3):
    row = cmk(a,b,c)
    print(f'{k[row]} = {tmp[row]};')
```

$\bm m=K^{-1}\bm\kappa$

```python
Ki = K.inv()
Ki
```

```python
tmp = Ki@k
for q in range(Q):
    tmp[q] = sp.simplify(sp.collect(sp.collect(tmp[q], u), v))
for a,b,c in it.product(range(0,3), repeat=3):
    row = cmk(a,b,c)
    print(f'{m[row]} = {tmp[row]};')
```
