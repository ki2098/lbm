```python
import sympy as sp
import numpy as np
import math
```

```python
def pdfid(i, j, k):
    return (i + 1)*9 + (j + 1)*3 + (k + 1)

def cmkid(a, b, c):
    return a*9 + b*3 + c
```

```python
f = np.ndarray(27, dtype=object)
for i in range(27):
    f[i] = sp.Symbol(f'f[{i}]')

m = np.ndarray(27, dtype=object)
for i in range(27):
    m[i] = sp.Symbol(f'm[{i}]')
```

```python
M = np.zeros((27, 27), dtype=int)
for a in range(3):
    for b in range(3):
        for c in range(3):
            for i in {-1,0,1}:
                for j in {-1,0,1}:
                    for k in {-1,0,1}:
                        M[cmkid(a,b,c), pdfid(i,j,k)] = (i**a)*(j**b)*(k**c)

M = sp.Matrix(M)
M
```

```python
Mi = M.inv()
Mi
```

```python
M@Mi
```

```python
M@f
```

```python
np.asarray(sp.factor(Mi@m))
```
