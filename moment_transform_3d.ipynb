```python
import numpy as np
import sympy as sp
import math
import itertools as it
```

```python
def cmk(i,j,k):
    return i*9 + j*3 + k

def pdf(i,j,k):
    return (i + 1)*9 + (j + 1)*3 + (k + 1)
```

# 1. Transform between PDFs and raw moments

Matrix $M$ that performs $\bm m=M\bm f$

```python
M = np.ndarray((27,27), dtype=int)
for a0,a1,a2 in it.product({0,1,2}, repeat=3):
    for i0,i1,i2 in it.product({-1,0,1}, repeat=3):
        rowid = cmk(a0,a1,a2)
        colid = pdf(i0,i1,i2)
        M[rowid, colid] = (i0**a0) * (i1**a1) * (i2**a2)

M
```

```python
Mi = np.linalg.inv(M)
Mi
```

```python
f = np.ndarray(27, dtype=object)
for i in range(27):
    f[i] = sp.Symbol(f'f[{i}]')
```

$\bm m = M\bm f$

```python
tmp = M@f
for i,row in enumerate(tmp):
    print(f'm[{i}] = {row}')
```

```python
m = np.ndarray(27, dtype=object)
for i in range(27):
    m[i] = sp.Symbol(f'm[{i}]')
```

$\bm f = M^{-1}\bm m$

```python
tmp = sp.factor(Mi@m)
for i,row in enumerate(tmp):
    print(f'm[{i}] = {row}')
```

# 2. Transform between raw and central moments

Matrix $K$ that performs $\bm\kappa=K\bm m$

```python
u, v, w = sp.symbols('u v w')
K = np.ndarray((27,27), dtype=object)
for a0,a1,a2 in it.product({0,1,2}, repeat=3):
    for i0,i1,i2 in it.product({0,1,2}, repeat=3):
        rowid = cmk(a0,a1,a2)
        colid = cmk(i0,i1,i2)
        K[rowid, colid] = \
            math.comb(a0,i0) * math.comb(a1,i1) * math.comb(a2,i2) \
        *   ((-u)**(a0-i0)) * ((-v)**(a1-i1)) * ((-w)**(a2-i2))

K = sp.Matrix(K)
K
```

```python
Ki = K.inv()
Ki
```

```python
K@Ki
```

```python
k = np.ndarray(27, dtype=object)
for i in range(27):
    k[i] = sp.Symbol(f'k[{i}]')
```

$\bm\kappa=K\bm m$

```python
tmp = K@k
for i,row in enumerate(tmp):
    tmp[i] = sp.simplify(sp.collect(sp.collect(row, u), v))
for i,row in enumerate(tmp):
    print(f'm[{i}] = {row}')
```

$\bm m = K^{-1}\bm\kappa$

```python
tmp = Ki@m
for i,row in enumerate(tmp):
    tmp[i] = sp.simplify(sp.collect(sp.collect(row, u), v))
for i,row in enumerate(tmp):
    print(f'k[{i}] = {row}')
```
